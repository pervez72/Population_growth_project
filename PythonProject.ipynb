### Density and Population Growth of some years

```python
df["Difference"] = df["2023"] - df["2000"]
df["Density23"] = df["2023"] / df["Area (km2)"]

print(df.columns)
```

### Top 10 Highest Countries by Population in 2023

```python
top_10_countries = df.nlargest(10, "2023")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Country Name", y="2023")
plt.title("Top 10 Highest Countries by Population")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.xticks(rotation=45)
plt.savefig("graph1.jpg", format="jpg", dpi=300)
plt.show()
```

### Top 10 Lowest Countries by Population in 2023 

```python
top_10_countries = df.nsmallest(10, "2023")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Country Name", y="2023")
plt.title("Top 10 Lowest Countries by Population")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.savefig("graph2.jpg", format="jpg", dpi=300)
plt.show()
```

### Top 10 Highest Countries by Population Growth from 2000 to 2023  
in x axis: population (in crore).  
in y axis: country name.

```python
top_10_countries = df.nlargest(10, "Difference")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Difference", y="Country Name")
plt.title("Top 10 Highest Countries by Population Growth from 2000 to 2023")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.savefig("graph3.jpg", format="jpg", dpi=300)
plt.show()
```

### Top 10 Lowest Countries by Population Growth from 2000 to 2023

```python
top_10_countries = df.nsmallest(10, "Difference")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Difference", y="Country Name")
plt.title("Top 10 Lowest Countries by Population Growth from 2000 to 2023")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.savefig("graph4.jpg", format="jpg", dpi=300)
plt.show()
```

### Population Density (per km2) : Highest 10 Country in 2023

```python
top_10_countries = df.nlargest(10, "Density23")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Country Name", y="Density23", color='skyblue')
plt.xlabel('Country Name')
plt.ylabel('Population Density (people per sq km)')
plt.title('Population Density : Highest 10 Country')
plt.xticks(rotation=45)
plt.savefig("graph5.jpg", format="jpg", dpi=300)
plt.show()
```

### Population Density (per km2) : Lowest 10 Country in 2023

```python
top_10_countries = df.nsmallest(10, "Density23")

plt.figure(figsize=(8, 4))
sns.barplot(data=top_10_countries, x="Country Name", y="Density23", color='green')
plt.xlabel('Country Name')
plt.ylabel('Population Density (people per sq km)')
plt.title('Population Density : Lowest 10 Country')
plt.xticks(rotation=45)
plt.savefig("graph6.jpg", format="jpg", dpi=300)
plt.show()
```
