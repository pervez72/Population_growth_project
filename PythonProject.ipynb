```python
sns.histplot(df['1960'], bins=5, kde=True)
plt.title("Population of 2023")
plt.xlabel("Population")
plt.ylabel("frequency")
plt.show()
```

```python
df.sort_values(by="2023", ascending=False).head(10)
```

```python
df["Diff 2000to2023"] = df["2023"]-df["2000"]
print(df.columns)
```

```python
# Sort and select top 10 countries
top_10_countries = df.nlargest(10, "2023")

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_countries, x="2023", y="Country Name",color='red')
plt.title("Top 10 Highest Countries by Population")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.savefig("paichart.png",transparent=True,pad_inch=2)

# Show the plot
plt.show();
```

```python
# Sort and select top 10 countries
top_10_countries = df.nsmallest(10, "2023")

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_countries, x="2023", y="Country Name")
plt.title("Top 10 Lowest Countries by Population")
plt.xlabel("Population")
plt.ylabel("Country Name")
plt.savefig("bar1.png",transparent=True,pad_inch=2)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = {
    'country': ['USA', 'India', 'China', 'Germany', 'Brazil'],
    'population': [331, 1391, 1441, 83, 213]  # Population in millions
}

# Create a DataFrame
df = pd.DataFrame(data)

# Filter specific countries
selected_countries = ['USA', 'India', 'China']
filtered_df = df[df['country'].isin(selected_countries)]

# Plot a bar chart
plt.figure(figsize=(8, 5))
plt.bar(filtered_df['country'], filtered_df['population'], color='orange')

# Add titles and labels
plt.title('Population of Selected Countries', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Population (millions)', fontsize=14)

# Show the plot
plt.show()
```
